# Inline Keyboard的應用

如何傳送及包裝一個Inline Keyboard？
* bot.sendMessage <- ( reply_markup ) InlineKeyboardMarkup <- ( inline_keyboard ) InlineKeyboardButton

```python
def make_kb_with_list(L):
    myBoard = []
    for item in L:
        myBoard.append([InlineKeyboardButton(
                        text=item,
                        callback_data=item
                    )])
    return InlineKeyboardMarkup(inline_keyboard=myBoard)
```

```python
def print_msg(msg):
    print(json.dumps(msg, indent=4))

def on_chat(msg):
    header = telepot.glance(msg, flavor="chat")

    print_msg(msg)
    user = header[2] #指定成易讀變數
    if header[0] == "text":
        text = msg["text"]
        # 遇到 command 
        if text.startswith("/"):
            command = text.lstrip("/")
            
            if command == "start":
                text = "Hi {}, 我是Cookbot 我會教你煮飯"
                bot.sendMessage(user, text.format(msg["from"]["first_name"]))
                bot.sendMessage(user,'{}, 你今天要處理冰箱的哪個主要食材呢？'.format(msg["from"]["first_name"]))
            
        # other msg
        elif text == "蛋": #食材在範圍列表內
            bot.sendMessage(user,'{}，我覺der可以，讓我來告訴你如何料理! '.format(text))
            bot.sendMessage(user, '你有這些食材可以搭配嗎？')
#                 bot.sendMessage(user, '你有這些食材嗎？:{}'.format(related_ingredient(text))) 
#             bot.sendMessage(user, related_ingredient(text))
#             bot.sendMessage(user, make_kb_with_list(related_ingredient(text)))
            bot.sendMessage(user, text="選擇回應", reply_markup=replyKeyboard)
            #bot.sendMessage(user,ingredients[text])
        elif text == "1": #番茄炒蛋
            bot.sendMessage(user,"https://www.ytower.com.tw/recipe/iframe-recipe.asp?seq=C01-0476")
            
        elif text == "2": #番茄牛肉燴飯
            bot.sendMessage(user,"https://www.ytower.com.tw/recipe/iframe-recipe.asp?seq=G01-345")
        else: #食材不在範圍內
            bot.sendMessage(user, '抱歉，連我也不知道{} 要怎麼料理 > <'.format(text))
            image_url = "http://img.memecdn.com/i-cant-cook-qnq_c_3961675.jpg" #when I cook@@s
            bot.sendPhoto(user, image_url)
            bot.sendMessage(user, "試試其他食材？")


    
#回應按鈕
replyKeyboard = InlineKeyboardMarkup(inline_keyboard=[
                [
                    InlineKeyboardButton(
                        text="烏龍麵",
                        callback_data="烏龍麵"
                    )   
                ],
                [
                    InlineKeyboardButton(
                        text="九層塔",
                        callback_data="九層塔"
                    )
                    
                ],
                [
                    InlineKeyboardButton(
                        text="牛肉",
                        callback_data="牛肉"
                    )
                ],
                [
                    InlineKeyboardButton(
                        text="番茄",
                        callback_data="番茄"
                    )
                ]
    
            ])           
def on_callback_query(msg):

    query_id, from_id, query_data = telepot.glance(msg, flavor='callback_query')
    print('Callback Query:', query_id, from_id, query_data)
    bot.answerCallbackQuery(query_id, '太好了，你有{}'.format(query_data))
    bot.sendMessage(from_id,'以下是番茄+蛋的料理...')
    bot.sendMessage(from_id,"1:番茄炒蛋（食材齊全） \n2:番茄牛肉燴飯（缺 牛肉 ） \n \n 請回覆代號")
```
